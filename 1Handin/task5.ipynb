```python
#Opgave 5 : Lav en metode der optegner alle monumenter på kortet ved brug af plotting
#Opgave 6 : Lav interaktion de plottede monumenter med visning af navn

import pandas as pd
import folium
import task3
import task1
import numpy as np

'''Denne viser alle monumenter på kort ved brug af plotting'''

def plotting_monuments():
    file="monumenter.csv"
    df = pd.read_csv(file)
    df_filter = df.dropna()
    df2 = df_filter
    map_monuments = folium.Map(location=(55.691976031093, 12.549923916910467), zoom_start=12)#kooridinater for Assistens Kirkegården fra Google maps
    for index, row in df2.iterrows():
        cooridinates = task3.cooridinates_of_monument(file, id=row['id'])
        tooltip = "Click Here For More Info"
        marker = folium.Marker(
            location=[cooridinates['latitude'], cooridinates['longitude']],
            popup="<strong>"+str(row['navn'])+"</strong>",
           tooltip=tooltip)
        marker.add_to(map_monuments)
    return map_monuments

plotting_monuments()
```
